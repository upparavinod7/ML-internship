## 15.1 Retrieval-Augmented Generation (RAG) & Vector Databases
RAG combines LLMs with an external knowledge base to provide context-aware, accurate answers.
We implement a simplified in-memory vector database and cosine similarity matching to simulate RAG.



````python
# Chapter 15.1: Simplified Vector Search (FAISS style logic)
import numpy as np

# 1. Mock Knowledge Base & Embeddings
documents = [
    "Machine Learning algorithms learn from data features.",
    "Deep Learning relies on multi-layer artificial neural networks.",
    "RAG improves LLM answers by supplying external documents.",
    "Docker containers package application code and dependencies."
]

# Simple mock embedding generator: TF-IDF style vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
doc_embeddings = vectorizer.fit_transform(documents).toarray()

# 2. Query processing
query = "Tell me about RAG and LLM context"
query_embedding = vectorizer.transform([query]).toarray()

# 3. Vector Similarity search: Cosine Similarity
# similarity = dot(A, B) / (norm(A) * norm(B))
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2.T)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2, axis=1)
    return (dot_product / (norm_v1 * norm_v2)).flatten()

similarities = cosine_similarity(query_embedding, doc_embeddings)
best_match_idx = np.argmax(similarities)

print("Query:", query)
print(f"Best Matched Document (Score: {similarities[best_match_idx]:.4f}):")
print(documents[best_match_idx])

````

## 15.2 LangChain & Agents
Agents use LLMs as reasoning engines to decide which tools to execute.



````python
# Chapter 15.2: Mock Agent Tool Execution Loop
class Tool:
    def __init__(self, name, func):
        self.name = name
        self.func = func

def search_web(query):
    return f"Search result for '{query}': DeepMind's Gemini models excel at multimodal tasks."

def calculate(expression):
    try:
        return str(eval(expression))
    except Exception as e:
        return f"Error: {e}"

tools = {
    "web_search": Tool("web_search", search_web),
    "calculator": Tool("calculator", calculate)
}

# Simulated LLM Reasoning output
# The LLM decides to call the calculator, then use that to answer.
mock_llm_steps = [
    {"thought": "I need to calculate 25 * 40 first.", "action": "calculator", "action_input": "25 * 40"},
    {"thought": "I have the calculation result. Now I can formulate the final answer.", "final_answer": "The result of 25 times 40 is 1000."}
]

for step in mock_llm_steps:
    if "action" in step:
        tool_name = step["action"]
        tool_input = step["action_input"]
        print(f"Thought: {step['thought']}")
        print(f"Calling tool: {tool_name} with input: {tool_input}")
        result = tools[tool_name].func(tool_input)
        print(f"Tool Output: {result}")
    elif "final_answer" in step:
        print(f"Final Answer: {step['final_answer']}")

````

# 🚀 Chapter 16: Deployment & MLOps

Model deployment transitions a trained model from development to production.

## 16.1 FastAPI Model Serving
We define a mock FastAPI app structure to serve model predictions.



````python
# Chapter 16.1: FastAPI Model Endpoint Simulation
# We write python code representing how to define the API routing.
from pydantic import BaseModel

# 1. Define Request / Response schemas
class PredictionRequest(BaseModel):
    experience_years: float
    education_level: str

class PredictionResponse(BaseModel):
    estimated_salary: float
    status: str

# 2. Mock endpoint handler
def predict_salary_endpoint(request: PredictionRequest):
    # Simulated model logic: Base 40k + 5k/year experience
    edu_bonus = 10000 if request.education_level.lower() == "masters" else 0
    salary = 40000 + (request.experience_years * 5000) + edu_bonus
    return PredictionResponse(estimated_salary=salary, status="success")

# Test prediction handler logic
sample_req = PredictionRequest(experience_years=5.5, education_level="masters")
response = predict_salary_endpoint(sample_req)
print(f"Endpoint Request: {sample_req.model_dump()}")
print(f"Endpoint Response: {response.model_dump()}")

````
