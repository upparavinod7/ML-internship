## Pandas: A Powerful Data Manipulation Library in Python

Pandas is an open-source Python library widely used for data manipulation and analysis. It provides high-performance, easy-to-use data structures and data analysis tools. The two primary data structures in pandas are Series and DataFrames.

### 1. Series

A `Series` is a one-dimensional labeled array capable of holding any data type (integers, strings, floats, Python objects, etc.). It is similar to a column in a spreadsheet or a SQL table, or a dictionary in Python.

- **Creation**: A Series can be created from a list, numpy array, or dictionary.
- **Indexing**: Each element in a Series has a label, or index, which can be an integer or a custom label (like strings).
- **Characteristics**:
    - **Homogeneous data**: All elements in a Series usually have the same data type.
    - **Size-immutable**: Once created, the size of a Series cannot be changed without creating a new one.
    - **Labeled index**: Allows for flexible data access and alignment.

**Example:**
```python
import pandas as pd
import numpy as np

# From a list
s1 = pd.Series([1, 2, 3, 4])
# From a numpy array
s2 = pd.Series(np.array(['a', 'b', 'c']))
# With custom index
s3 = pd.Series([10, 20, 30], index=['x', 'y', 'z'])

print("Series from list:\n", s1)
print("\nSeries from numpy array:\n", s2)
print("\nSeries with custom index:\n", s3)
```

### 2. DataFrame

A `DataFrame` is a two-dimensional labeled data structure with columns of potentially different types. It is similar to a spreadsheet, a SQL table, or a dictionary of Series objects. It is the most commonly used pandas object.

- **Creation**: DataFrames can be created in various ways:
    - From dictionaries of Series, lists, or numpy arrays.
    - From a list of dictionaries.
    - From structured data files (CSV, Excel, SQL databases).
- **Indexing**: It has both a row and column index.
- **Characteristics**:
    - **Heterogeneous data**: Columns can have different data types.
    - **Size-mutable**: Columns can be inserted and deleted.
    - **Labeled axes**: Rows and columns have labels (indices).

**Example:**
```python
import pandas as pd

# From a dictionary of lists
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}
df1 = pd.DataFrame(data)

# From a list of dictionaries
data_list = [
    {'Name': 'David', 'Age': 40, 'City': 'Houston'},
    {'Name': 'Eve', 'Age': 28, 'City': 'Miami'}
]
df2 = pd.DataFrame(data_list)

print("DataFrame from dictionary:\n", df1)
print("\nDataFrame from list of dictionaries:\n", df2)
```

### Key Pandas Operations and Functionalities

#### A. Data Inspection

- `df.head(n)`: Returns the first `n` rows of the DataFrame.
- `df.tail(n)`: Returns the last `n` rows of the DataFrame.
- `df.info()`: Prints a concise summary of a DataFrame, including data types and non-null values.
- `df.describe()`: Generates descriptive statistics of numerical columns (count, mean, std, min, max, quartiles).
- `df.shape`: Returns a tuple representing the dimensionality of the DataFrame (rows, columns).
- `df.columns`: Returns a list of column names.
- `df.index`: Returns the index (row labels) of the DataFrame.
- `df.dtypes`: Returns the data type of each column.

#### B. Selection and Indexing

- `df['column_name']`: Select a single column (returns a Series).
- `df[['col1', 'col2']]`: Select multiple columns (returns a DataFrame).
- `df.loc[row_label, col_label]`: Label-based indexing for rows and columns.
- `df.iloc[row_position, col_position]`: Integer-position-based indexing for rows and columns.
- **Boolean Indexing**: Filtering rows based on conditions.
    ```python
    df[df['Age'] > 30]
    ```

#### C. Data Cleaning

- `df.dropna()`: Removes rows with missing values.
- `df.fillna(value)`: Fills missing values with a specified `value`.
- `df.duplicated()`: Returns a boolean Series indicating duplicate rows.
- `df.drop_duplicates()`: Removes duplicate rows.
- `df.astype(dtype)`: Converts a column to a specific data type.
- `pd.to_datetime()`: Converts a Series or column to datetime objects.

#### D. Data Manipulation

- `df.apply(func, axis=0/1)`: Applies a function along an axis of the DataFrame.
- `df.groupby('column_name')`: Groups rows based on the values in a column, often followed by aggregation functions (e.g., `.sum()`, `.mean()`, `.count()`).
- `df.merge()`: Combines DataFrames based on common columns or indices (similar to SQL JOINs).
- `df.concat()`: Stacks DataFrames along an axis (rows or columns).
- `df.pivot_table()`: Creates a spreadsheet-style pivot table.
- `df.sort_values(by='column_name', ascending=True/False)`: Sorts a DataFrame by column values.
- `df.set_index('column_name')`: Sets a column as the DataFrame index.
- `df.reset_index()`: Resets the index to the default integer index.

#### E. Input/Output Operations

- `pd.read_csv('file.csv')`: Reads data from a CSV file into a DataFrame.
- `df.to_csv('output.csv', index=False)`: Writes a DataFrame to a CSV file.
- `pd.read_excel('file.xlsx')`: Reads data from an Excel file.
- `df.to_excel('output.xlsx', index=False)`: Writes a DataFrame to an Excel file.
- `pd.read_sql('SELECT * FROM table', connection)`: Reads data from a SQL query or table.

## Deep Dive into Pandas Series

A Pandas Series is a one-dimensional labeled array. Think of it as a single column of data with a corresponding index. It can hold data of any type (integers, floats, strings, Python objects, etc.).

### 1. Series Creation

A Series can be created from various data structures:

- **From a Python list**: Simple and straightforward.
- **From a NumPy array**: Useful when you already have numerical data in array format.
- **From a Python dictionary**: The keys of the dictionary become the index, and the values become the data.
- **With custom index**: You can explicitly specify the index labels during creation.
- **Scalar value**: You can create a Series with a single scalar value, which will be repeated for a specified index length.

````python
import pandas as pd
import numpy as np

print("--- Creating Series from a list ---")
s_list = pd.Series([10, 20, 30, 40, 50])
print("Series from list:\n", s_list)

print("\n--- Creating Series from a NumPy array ---")
s_np = pd.Series(np.array([1.1, 2.2, 3.3, 4.4]))
print("Series from NumPy array:\n", s_np)

print("\n--- Creating Series from a dictionary ---")
dict_data = {'a': 100, 'b': 200, 'c': 300}
s_dict = pd.Series(dict_data)
print("Series from dictionary:\n", s_dict)

print("\n--- Creating Series with a custom index ---")
s_custom_idx = pd.Series([5, 6, 7], index=['apple', 'banana', 'cherry'])
print("Series with custom index:\n", s_custom_idx)

print("\n--- Creating Series from a scalar value with an index ---")
s_scalar = pd.Series(5, index=[0, 1, 2, 3])
print("Series from scalar:\n", s_scalar)
````

### 2. Series Indexing and Selection

Accessing elements in a Series is similar to NumPy arrays or Python dictionaries, but with more flexibility due to its labeled index.

- **Positional indexing**: Access elements using their integer position (0-based).
- **Label-based indexing**: Access elements using their assigned labels.
- **Slicing**: Select a range of elements using both positional and label-based slicing.
- **Boolean indexing**: Filter elements based on a condition.

````python
import pandas as pd

my_series = pd.Series([10, 20, 30, 40, 50], index=['A', 'B', 'C', 'D', 'E'])
print("Original Series:\n", my_series)

print("\n--- Positional Indexing ---")
print("Element at position 0:", my_series[0])
print("Element at position 3:", my_series[3])

print("\n--- Label-based Indexing ---")
print("Element with label 'A':", my_series['A'])
print("Element with label 'D':", my_series['D'])

print("\n--- Slicing (Positional) ---")
print("Elements from position 1 to 3 (exclusive):\n", my_series[1:4])

print("\n--- Slicing (Label-based) ---")
# Note: Label-based slicing is inclusive of the end label
print("Elements from 'B' to 'D' (inclusive):\n", my_series['B':'D'])

print("\n--- Boolean Indexing ---")
# Select elements greater than 30
print("Elements > 30:\n", my_series[my_series > 30])

# Select elements with labels 'A' and 'C'
print("Elements with labels 'A' and 'C':\n", my_series[['A', 'C']])
````

### 3. Series Properties and Methods

Series objects come with various useful attributes and methods for inspection and basic operations.

- `.index`: The index (labels) of the Series.
- `.values`: The actual data values as a NumPy array.
- `.dtype`: The data type of the elements in the Series.
- `.name`: An optional name for the Series.
- `.size`: The number of elements in the Series.
- `.shape`: The shape of the Series (always a tuple with one element for 1D).
- `.unique()`: Returns unique values in the Series.
- `.value_counts()`: Returns a Series containing counts of unique values.
- `.head(n)` / `.tail(n)`: Returns the first/last `n` elements.
- Basic arithmetic operations (+, -, *, /) are applied element-wise.

````python
import pandas as pd
import numpy as np

# Create a sample Series for demonstration
data_values = [10, 20, 10, 30, 20, 40, 10, np.nan]
labels = ['alpha', 'beta', 'gamma', 'delta', 'epsilon', 'zeta', 'eta', 'theta']
my_series = pd.Series(data_values, index=labels, name='SampleData')

print("Original Series:\n", my_series)

print("\n--- Properties ---")
print("Index:", my_series.index)
print("Values (as NumPy array):", my_series.values)
print("Data Type:", my_series.dtype)
print("Name:", my_series.name)
print("Size (number of elements):", my_series.size)
print("Shape:", my_series.shape)

print("\n--- Methods ---")
print("First 3 elements (head):\n", my_series.head(3))
print("\nLast 2 elements (tail):\n", my_series.tail(2))

print("\nUnique values:", my_series.unique())

print("\nValue Counts:\n", my_series.value_counts())

print("\n--- Arithmetic Operations ---")
print("Series + 5:\n", my_series + 5)
print("Series * 2:\n", my_series * 2)

# Filling missing values for demonstration
print("\nSeries after filling NaN with mean:\n", my_series.fillna(my_series.mean()))
````

### 4. Visualizing Series Data

While Series are 1D, basic plots can help understand their distribution or trends, especially for numerical data. `matplotlib` and `seaborn` are commonly used for visualization.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a Series with some random data
np.random.seed(42)
data_for_plot = pd.Series(np.random.rand(20) * 100, name='Random Values')

# Create a Series with categorical data for value_counts visualization
category_data = pd.Series(['A', 'B', 'A', 'C', 'B', 'A', 'D', 'C', 'B', 'A', 'D'], name='Categories')


print("--- Basic Plot (Line Plot) ---")
plt.figure(figsize=(10, 4))
data_for_plot.plot(title='Line Plot of Random Values')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n--- Histogram (Distribution) ---")
plt.figure(figsize=(10, 4))
data_for_plot.plot(kind='hist', bins=5, edgecolor='black', title='Histogram of Random Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

print("\n--- Bar Plot of Value Counts (Categorical Data) ---")
plt.figure(figsize=(8, 5))
category_data.value_counts().plot(kind='bar', title='Count of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
````

### 1.1 Creating a DataFrame from a Dictionary of Lists or NumPy Arrays

### 1.3 Creating a DataFrame from a Dictionary of Series

Each value in the dictionary can also be a pandas Series. This is useful for combining existing Series objects into a DataFrame. Pandas will align the data based on the Series' indices.

````python
import pandas as pd

print("--- Creating DataFrame from a dictionary of Series ---")
s_names = pd.Series(['Harry', 'Ivy', 'Jack'], index=['idx1', 'idx2', 'idx3'])
s_ages = pd.Series([22, 29, 50], index=['idx2', 'idx1', 'idx3'])
s_cities = pd.Series(['London', 'Paris', 'Tokyo'], index=['idx1', 'idx2', 'idx3'])

data_dict_series = {
    'Name': s_names,
    'Age': s_ages,
    'City': s_cities
}
df_from_dict_series = pd.DataFrame(data_dict_series)
print("DataFrame:\n")
display(df_from_dict_series)

print("\nNotice how 'Age' for 'idx1' and 'idx2' are aligned by index, and NaN is introduced where an index is missing.")
````

### 2. DataFrame Indexing and Selection

Selecting data from a DataFrame is a fundamental operation. Pandas offers powerful and flexible ways to do this.

#### 2.1 Selecting Columns

-   **Single Column**: Select a single column using bracket notation (`df['column_name']`). This returns a Pandas Series.
-   **Multiple Columns**: Select multiple columns by passing a list of column names (`df[['col1', 'col2']]`). This returns a DataFrame.

````python
import pandas as pd

df = pd.DataFrame({
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    'Price': [1200, 25, 75, 300],
    'Quantity': [10, 50, 30, 15]
})

print("Original DataFrame:\n")
display(df)

print("\n--- Selecting a single column ('Product') ---")
product_series = df['Product']
display(product_series)
print("Type of result:", type(product_series))

print("\n--- Selecting multiple columns ('Product' and 'Price') ---")
product_price_df = df[['Product', 'Price']]
display(product_price_df)
print("Type of result:", type(product_price_df))
````

#### 2.2 Selecting Rows with `.loc` and `.iloc`

-   **`.loc` (Label-based indexing)**: Used for selecting data by explicit label (row and column names).
-   **`.iloc` (Integer-location based indexing)**: Used for selecting data by integer position (0-based index of rows and columns).

````python
import pandas as pd

df_loc_iloc = pd.DataFrame(
    np.random.randint(10, 100, size=(5, 4)),
    index=['row_a', 'row_b', 'row_c', 'row_d', 'row_e'],
    columns=['col_w', 'col_x', 'col_y', 'col_z']
)

print("Original DataFrame:\n")
display(df_loc_iloc)

print("\n--- Using .loc (label-based) ---")
print("Select 'row_b':\n", df_loc_iloc.loc['row_b'])
print("\nSelect 'row_a' and 'col_x':", df_loc_iloc.loc['row_a', 'col_x'])
print("\nSelect multiple rows and columns:\n", df_loc_iloc.loc[['row_a', 'row_c'], ['col_w', 'col_y']])
print("\nSelect all columns for 'row_d' to 'row_e' (inclusive):\n", df_loc_iloc.loc['row_d':'row_e', :])

print("\n--- Using .iloc (integer-location based) ---")
print("Select row at index 0:\n", df_loc_iloc.iloc[0])
print("\nSelect element at row 1, column 2:", df_loc_iloc.iloc[1, 2])
print("\nSelect multiple rows and columns by position:\n", df_loc_iloc.iloc[[0, 2], [0, 3]])
print("\nSelect rows from index 1 to 3 (exclusive) and all columns:\n", df_loc_iloc.iloc[1:4, :])
````

#### 2.3 Boolean Indexing (Filtering Data)

Boolean indexing allows you to select rows based on a condition applied to one or more columns. It's a very powerful way to filter your data.

````python
import pandas as pd

df_filter = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [85, 92, 78, 95, 88],
    'Grade': ['B', 'A', 'C', 'A', 'B']
})

print("Original DataFrame:\n")
display(df_filter)

print("\n--- Students with Score > 85 ---")
filtered_by_score = df_filter[df_filter['Score'] > 85]
display(filtered_by_score)

print("\n--- Students with Grade 'A' ---")
filtered_by_grade = df_filter[df_filter['Grade'] == 'A']
display(filtered_by_grade)

print("\n--- Students with Score > 80 AND Grade 'B' ---")
filtered_complex = df_filter[(df_filter['Score'] > 80) & (df_filter['Grade'] == 'B')]
display(filtered_complex)

print("\n--- Students with Grade 'A' OR 'C' ---")
filtered_or = df_filter[df_filter['Grade'].isin(['A', 'C'])]
display(filtered_or)
````

### 4. DataFrame Operations (Arithmetic)

Pandas DataFrames support element-wise arithmetic operations, similar to NumPy arrays. When performing operations between DataFrames or a DataFrame and a Series, pandas aligns the data based on indices and column names.

````python
import pandas as pd

df_ops1 = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6]
})
df_ops2 = pd.DataFrame({
    'A': [7, 8, 9],
    'B': [10, 11, 12]
})

print("DataFrame 1:\n")
display(df_ops1)
print("\nDataFrame 2:\n")
display(df_ops2)

print("\n--- Addition (df_ops1 + df_ops2) ---")
display(df_ops1 + df_ops2)

print("\n--- Subtraction (df_ops1 - df_ops2) ---")
display(df_ops1 - df_ops2)

print("\n--- Multiplication (df_ops1 * 2) ---")
display(df_ops1 * 2)

print("\n--- Division (df_ops1 / df_ops2) ---")
display(df_ops1 / df_ops2)

print("\n--- Operations with a Series (broadcasts row-wise) ---")
series_add = pd.Series([100, 200], index=['A', 'B'])
print("Series for addition:\n", series_add)
display(df_ops1 + series_add)
````

### 5. Data Cleaning

Real-world data is often messy and incomplete. Pandas provides essential tools for cleaning and preprocessing data.

#### 5.1 Handling Missing Values

Missing values (often represented as `NaN` or `None`) are common. Identifying and handling them is crucial.

-   `.isnull()` / `.isna()`: Returns a boolean DataFrame indicating where values are missing.
-   `.notnull()` / `.notna()`: Returns a boolean DataFrame indicating where values are not missing.
-   `.dropna()`: Removes rows or columns containing missing values.
-   `.fillna()`: Fills missing values with a specified value or method.

````python
import pandas as pd
import numpy as np

df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [6, np.nan, 8, 9, 10],
    'C': ['x', 'y', 'z', np.nan, 'w']
})

print("Original DataFrame with missing values:\n")
display(df_missing)

print("\n--- .isnull(): Identify missing values ---")
display(df_missing.isnull())

print("\n--- .dropna(): Drop rows with any missing values ---")
display(df_missing.dropna())

print("\n--- .dropna(how='all'): Drop rows where ALL values are missing ---")
# Add a row with all NaNs for demonstration
df_missing_all = df_missing.append(pd.Series([np.nan, np.nan, np.nan], index=['A', 'B', 'C']), ignore_index=True)
display(df_missing_all.dropna(how='all'))

print("\n--- .fillna(0): Fill missing values with 0 ---")
display(df_missing.fillna(0))

print("\n--- .fillna(method='ffill'): Fill missing values using forward fill ---")
display(df_missing.fillna(method='ffill'))

print("\n--- .fillna(df.mean()): Fill missing numerical values with column mean ---")
df_filled_mean = df_missing.copy()
df_filled_mean['A'] = df_filled_mean['A'].fillna(df_filled_mean['A'].mean())
df_filled_mean['B'] = df_filled_mean['B'].fillna(df_filled_mean['B'].mean())
display(df_filled_mean)
````

#### 5.2 Handling Duplicates

Duplicate rows can skew analysis. Pandas provides methods to identify and remove them.

-   `.duplicated()`: Returns a boolean Series indicating which rows are duplicates (defaults to keeping the first occurrence).
-   `.drop_duplicates()`: Removes duplicate rows.

````python
import pandas as pd

df_duplicates = pd.DataFrame({
    'ID': [1, 2, 1, 3, 2, 4],
    'Value': ['A', 'B', 'A', 'C', 'B', 'D']
})

print("Original DataFrame with duplicates:\n")
display(df_duplicates)

print("\n--- .duplicated(): Identify duplicate rows ---")
display(df_duplicates.duplicated())

print("\n--- .drop_duplicates(): Remove duplicate rows (keeps first by default) ---")
display(df_duplicates.drop_duplicates())

print("\n--- .drop_duplicates(keep='last'): Remove duplicate rows (keeps last) ---")
display(df_duplicates.drop_duplicates(keep='last'))

print("\n--- .drop_duplicates(subset=['ID']): Remove duplicates based on 'ID' column only ---")
display(df_duplicates.drop_duplicates(subset=['ID']))
````

#### 5.3 Changing Data Types (`.astype()`, `pd.to_datetime()`)

Correct data types are essential for proper analysis and efficient memory usage.

-   `.astype()`: Converts a column or DataFrame to a specified data type.
-   `pd.to_datetime()`: Specifically converts columns to datetime objects, which unlocks time-series functionalities.

````python
import pandas as pd

df_types = pd.DataFrame({
    'Number_Str': ['1', '2', '3', '4'],
    'Price_Float': [10.5, 20.0, 15.75, 12.0],
    'Date_Str': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04']
})

print("Original DataFrame dtypes:\n")
display(df_types.dtypes)

print("\n--- Convert 'Number_Str' to integer using .astype() ---")
df_types['Number_Str'] = df_types['Number_Str'].astype(int)
display(df_types.dtypes)

print("\n--- Convert 'Date_Str' to datetime objects using pd.to_datetime() ---")
df_types['Date_Str'] = pd.to_datetime(df_types['Date_Str'])
display(df_types.dtypes)

print("\n--- Accessing datetime properties (e.g., year, month, day) ---")
df_types['Year'] = df_types['Date_Str'].dt.year
df_types['Month'] = df_types['Date_Str'].dt.month
display(df_types)
````

### 6. Data Manipulation

Pandas provides a rich set of functions for transforming and reorganizing data.

#### 6.1 Applying Functions (`.apply()`, `.applymap()`, `.pipe()`)

-   `.apply()`: Apply a function along an axis (row or column) of the DataFrame. Can also be applied to a Series (a single column).
-   `.applymap()`: Apply a function element-wise to a DataFrame (useful for cell-level transformations). *Note: `applymap` is deprecated in newer pandas versions; use `df.map` for Series and `df.apply(lambda x: x.map(func))` or `df.transform` for DataFrames for similar element-wise operations.*
-   `.pipe()`: Allows for chaining operations with functions that take a DataFrame as their first argument.

````python
import pandas as pd

df_apply = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [100, 200, 300]
})

print("Original DataFrame:\n")
display(df_apply)

print("\n--- .apply() to a Series (column) to square values ---")
df_apply['A_squared'] = df_apply['A'].apply(lambda x: x**2)
display(df_apply)

print("\n--- .apply() to the DataFrame to sum each row ---")
df_apply['Row_Sum'] = df_apply.apply(lambda row: row['A'] + row['B'], axis=1)
display(df_apply)

print("\n--- .apply() to the DataFrame to find max of each column ---")
display(df_apply.apply(lambda col: col.max(), axis=0))

# Element-wise operation example (alternative to applymap)
print("\n--- Element-wise transformation (e.g., add 5 to each element) ---")
df_transformed = df_apply[['A', 'B']].apply(lambda x: x + 5)
display(df_transformed)
````

#### 6.2 Grouping Data (`.groupby()`)

`.groupby()` is one of the most powerful features of pandas, allowing you to group data based on one or more columns and then apply an aggregation function to each group (e.g., sum, mean, count, min, max).

````python
import pandas as pd

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'C', 'B', 'A'],
    'Value1': [10, 20, 15, 25, 30, 12],
    'Value2': [1, 2, 1, 3, 2, 1]
})

print("Original DataFrame:\n")
display(df_group)

print("\n--- Group by 'Category' and calculate mean of Value1 ---")
grouped_mean = df_group.groupby('Category')['Value1'].mean()
display(grouped_mean)

print("\n--- Group by 'Category' and get multiple aggregations ---")
grouped_agg = df_group.groupby('Category').agg(
    mean_value1=('Value1', 'mean'),
    sum_value2=('Value2', 'sum'),
    count=('Value1', 'count')
)
display(grouped_agg)

print("\n--- Group by multiple columns ---")
df_group_multi = pd.DataFrame({
    'Region': ['East', 'West', 'East', 'West', 'East'],
    'Product': ['A', 'A', 'B', 'B', 'A'],
    'Sales': [100, 150, 200, 120, 180]
})
display(df_group_multi)
print("\nGroup by Region and Product and sum Sales:\n")
display(df_group_multi.groupby(['Region', 'Product'])['Sales'].sum())
````

#### 6.3 Combining Data (`.merge()`, `.concat()`)

-   `.merge()`: Combines DataFrames based on common columns or indices (similar to SQL JOINs).
-   `.concat()`: Stacks DataFrames either row-wise (`axis=0`) or column-wise (`axis=1`).

````python
import pandas as pd

# Data for merging
df_customers = pd.DataFrame({
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})
df_orders = pd.DataFrame({
    'OrderID': [101, 102, 103, 104],
    'CustomerID': [2, 1, 4, 2],
    'Amount': [50, 75, 120, 60]
})

print("Customers DataFrame:\n")
display(df_customers)
print("\nOrders DataFrame:\n")
display(df_orders)

print("\n--- .merge(): Inner join on 'CustomerID' ---")
merged_df = pd.merge(df_customers, df_orders, on='CustomerID', how='inner')
display(merged_df)

print("\n--- .merge(): Left join to keep all customers ---")
left_merged_df = pd.merge(df_customers, df_orders, on='CustomerID', how='left')
display(left_merged_df)

# Data for concatenating
df_part1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_part2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df_part3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

print("\nDataFrame Part 1:\n")
display(df_part1)
print("\nDataFrame Part 2:\n")
display(df_part2)
print("\nDataFrame Part 3:\n")
display(df_part3)

print("\n--- .concat(axis=0): Stack DataFrames vertically ---")
concatenated_rows = pd.concat([df_part1, df_part2], axis=0, ignore_index=True)
display(concatenated_rows)

print("\n--- .concat(axis=1): Concatenate DataFrames horizontally ---")
concatenated_cols = pd.concat([df_part1, df_part3], axis=1)
display(concatenated_cols)
````

#### 6.4 Reshaping Data (`.pivot_table()`, `.stack()`, `.unstack()`)

Reshaping allows you to change the layout of your data, often to make it more suitable for specific analyses or visualizations.

-   `.pivot_table()`: Creates a spreadsheet-style pivot table as a DataFrame.
-   `.stack()`: "Stacks" the columns into rows, producing a Series or DataFrame with a MultiIndex.
-   `.unstack()`: "Unstacks" rows into columns, inverse of `stack()`.

````python
import pandas as pd

df_sales = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'Region': ['East', 'West', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 120, 180]
})

print("Original Sales DataFrame:\n")
display(df_sales)

print("\n--- .pivot_table(): Sum of sales by Date and Region, with Products as columns ---")
pivot_table_df = df_sales.pivot_table(values='Sales', index='Date', columns='Region', aggfunc='sum')
display(pivot_table_df)

# For stack/unstack, let's create a DataFrame with a MultiIndex
df_multi_index = df_sales.set_index(['Date', 'Region', 'Product'])
print("\nDataFrame with MultiIndex:\n")
display(df_multi_index)

print("\n--- .unstack(): Moves the innermost index level ('Product') to columns ---")
unstacked_df = df_multi_index.unstack(level='Product')
display(unstacked_df)

print("\n--- .stack(): Inverse of unstack, moves columns back to index ---")
stacked_series = unstacked_df.stack(level='Product')
display(stacked_series)
````

#### 6.5 Sorting Data (`.sort_values()`, `.sort_index()`)

Ordering your data is often necessary for analysis or presentation.

-   `.sort_values()`: Sorts the DataFrame by the values in one or more columns.
-   `.sort_index()`: Sorts the DataFrame by its index labels.

````python
import pandas as pd

df_sort = pd.DataFrame({
    'Name': ['Charlie', 'Alice', 'Bob', 'David'],
    'Age': [35, 25, 30, 28],
    'City': ['Chicago', 'New York', 'Los Angeles', 'Houston']
}, index=['C', 'A', 'B', 'D'])

print("Original DataFrame:\n")
display(df_sort)

print("\n--- .sort_values() by 'Age' (ascending by default) ---")
display(df_sort.sort_values(by='Age'))

print("\n--- .sort_values() by 'Name' (descending) ---")
display(df_sort.sort_values(by='Name', ascending=False))

print("\n--- .sort_index() (ascending by default) ---")
display(df_sort.sort_index())

print("\n--- .sort_index() (descending) ---")
display(df_sort.sort_index(ascending=False))
````

#### 6.6 Setting/Resetting Index (`.set_index()`, `.reset_index()`)

-   `.set_index()`: Sets one or more columns as the DataFrame index.
-   `.reset_index()`: Resets the index to the default integer index, moving the current index to a column (or dropping it).

````python
import pandas as pd

df_index = pd.DataFrame({
    'ID': [1, 2, 3],
    'Product': ['A', 'B', 'C'],
    'Price': [10, 20, 30]
})

print("Original DataFrame:\n")
display(df_index)

print("\n--- .set_index('ID'): Set 'ID' as the new index ---")
df_indexed = df_index.set_index('ID')
display(df_indexed)

print("\n--- .reset_index(): Reset the index, 'ID' becomes a column ---")
display(df_indexed.reset_index())

print("\n--- .reset_index(drop=True): Reset index and drop the old index ---")
display(df_indexed.reset_index(drop=True))
````

### 7. Input/Output Operations

Pandas makes it incredibly easy to read data from and write data to various file formats.

#### 7.1 Reading and Writing CSV Files (`pd.read_csv()`, `df.to_csv()`)

CSV (Comma Separated Values) is a very common plain-text format for tabular data.

````python
import pandas as pd
import os

# Create a sample DataFrame to write to CSV
df_to_csv = pd.DataFrame({
    'Col1': [1, 2, 3],
    'Col2': ['X', 'Y', 'Z']
})

# Define a filename
csv_filename = 'sample_data.csv'

print("Original DataFrame to be saved:\n")
display(df_to_csv)

print("\n--- .to_csv(): Writing DataFrame to CSV ---")
df_to_csv.to_csv(csv_filename, index=False) # index=False prevents writing the DataFrame index as a column
print(f"DataFrame saved to '{csv_filename}'")

print("\n--- pd.read_csv(): Reading DataFrame from CSV ---")
df_from_csv = pd.read_csv(csv_filename)
print("DataFrame read from CSV:\n")
display(df_from_csv)

# Clean up the created file
if os.path.exists(csv_filename):
    os.remove(csv_filename)
    print(f"Cleaned up: '{csv_filename}' removed.")
````

Excel files (`.xlsx`, `.xls`) are also widely used, especially in business contexts.

````python
import pandas as pd
import os

# Create a sample DataFrame to write to Excel
df_to_excel = pd.DataFrame({
    'Product': ['ItemA', 'ItemB', 'ItemC'],
    'Sales_Q1': [100, 150, 120],
    'Sales_Q2': [110, 140, 130]
})

# Define a filename
excel_filename = 'sample_data.xlsx'

print("Original DataFrame to be saved:\n")
display(df_to_excel)

print("\n--- .to_excel(): Writing DataFrame to Excel ---")
df_to_excel.to_excel(excel_filename, index=False, sheet_name='Quarterly_Sales')
print(f"DataFrame saved to '{excel_filename}'")

print("\n--- pd.read_excel(): Reading DataFrame from Excel ---")
df_from_excel = pd.read_excel(excel_filename, sheet_name='Quarterly_Sales')
print("DataFrame read from Excel:\n")
display(df_from_excel)

# Clean up the created file
if os.path.exists(excel_filename):
    os.remove(excel_filename)
    print(f"Cleaned up: '{excel_filename}' removed.")
````

### 8. Visualizing DataFrame Data

Pandas DataFrames integrate well with visualization libraries like Matplotlib and Seaborn, allowing you to plot data directly from your DataFrame to understand distributions, relationships, and trends.

-   `df.plot()`: A convenient wrapper around Matplotlib for quick plots.
-   `seaborn.scatterplot()`, `seaborn.histplot()`, etc.: For more sophisticated statistical plots.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a sample DataFrame for visualization
np.random.seed(42)
df_plot = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=10),
    'Sales': np.random.randint(100, 500, size=10) + np.arange(10) * 20,
    'Expenses': np.random.randint(50, 200, size=10) + np.arange(10) * 5,
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home Goods'], size=10)
})

print("Original DataFrame for Plotting:\n")
display(df_plot)

print("\n--- Line Plot of Sales Over Time (df.plot()) ---")
plt.figure(figsize=(10, 5))
df_plot.plot(x='Date', y='Sales', kind='line', title='Daily Sales Trend')
plt.ylabel('Sales Amount')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n--- Bar Plot of Category Counts (from Series.plot()) ---")
plt.figure(figsize=(8, 4))
df_plot['Category'].value_counts().plot(kind='bar', title='Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n--- Scatter Plot of Sales vs Expenses (Seaborn) ---")
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Expenses', y='Sales', hue='Category', data=df_plot, s=100)
plt.title('Sales vs Expenses by Category')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print("\n--- Histogram of Sales (Seaborn) ---")
plt.figure(figsize=(8, 5))
sns.histplot(df_plot['Sales'], bins=5, kde=True)
plt.title('Distribution of Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````
