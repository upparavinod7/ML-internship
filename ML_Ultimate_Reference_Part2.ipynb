# 📊 Chapter 5: Matplotlib Complete Guide

Matplotlib is a low-level plotting library in Python that provides total control over every element in a figure.

## 5.1 Basic Plot Types, Subplots, and Customization
We will generate standard plot types (Line, Scatter, Bar, Histogram) and structure them cleanly using `plt.subplots`.



```python
# Chapter 5.1: Matplotlib Essentials
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic dataset
x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Create a figure with 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 1. Line Plot (Sine and Cosine)
axs[0, 0].plot(x, y_sin, label='Sine', color='royalblue', linewidth=2)
axs[0, 0].plot(x, y_cos, label='Cosine', color='darkorange', linestyle='--', linewidth=2)
axs[0, 0].set_title('Line Plot (Sine & Cosine)')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].legend()
axs[0, 0].grid(True, linestyle=':', alpha=0.6)

# 2. Scatter Plot with colormap
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
axs[0, 1].scatter(x, y_sin + np.random.randn(100) * 0.2, c=colors, s=sizes, alpha=0.7, cmap='viridis')
axs[0, 1].set_title('Scatter Plot with Custom Scale')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')
axs[0, 1].grid(True)

# 3. Bar Chart
categories = ['Class A', 'Class B', 'Class C', 'Class D']
values = [25, 40, 15, 30]
axs[1, 0].bar(categories, values, color=['#4285F4', '#EA4335', '#FBBC05', '#34A853'], edgecolor='black')
axs[1, 0].set_title('Categorical Bar Chart')
axs[1, 0].set_ylabel('Performance Metric')

# 4. Histogram (Normal Distribution)
gaussian_data = np.random.randn(1000)
axs[1, 1].hist(gaussian_data, bins=30, color='teal', edgecolor='white', alpha=0.8)
axs[1, 1].set_title('Histogram of Gaussian Distribution')
axs[1, 1].set_xlabel('Value')
axs[1, 1].set_ylabel('Frequency')

# Global Layout adjustments
plt.tight_layout()
plt.show()

```

## 5.2 Advanced Customizations & Annotations
Adding annotations, legend fine-tuning, and coordinate highlighting.



```python
# Chapter 5.2: Annotations & Highlighting
fig, ax = plt.subplots(figsize=(8, 5))
x_vals = np.linspace(0, 5, 100)
y_vals = x_vals ** 2

ax.plot(x_vals, y_vals, color='purple', label='$y = x^2$')
ax.axvline(2.0, color='red', linestyle='--', alpha=0.5, label='Threshold X=2.0')
ax.axhline(4.0, color='green', linestyle=':', alpha=0.5, label='Threshold Y=4.0')

# Annotation
ax.annotate('Critical Point (2, 4)', xy=(2, 4), xytext=(3, 10),
            arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=6))

ax.set_title('Quadratic Growth with Reference Lines')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.legend(loc='upper left')
plt.show()

```

# 📈 Chapter 6: Seaborn Complete Guide

Seaborn is built on top of Matplotlib and integrates closely with Pandas data structures. It provides high-level APIs for drawing attractive statistical graphics.

## 6.1 Distribution & Categorical Plots
We use the synthetic tips dataset structure to explore Seaborn's capabilities.



```python
# Chapter 6.1: Seaborn Distributions & Categoricals
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate dummy dataset resembling customer tips
np.random.seed(42)
tips = pd.DataFrame({
    'total_bill': np.random.normal(20, 8, 200).clip(5, 50),
    'tip': np.random.normal(3, 1.5, 200).clip(1, 10),
    'sex': np.random.choice(['Male', 'Female'], 200),
    'smoker': np.random.choice(['Yes', 'No'], 200),
    'day': np.random.choice(['Thur', 'Fri', 'Sat', 'Sun'], 200),
    'time': np.random.choice(['Lunch', 'Dinner'], 200),
    'size': np.random.randint(1, 6, 200)
})

# Set Seaborn Style
sns.set_theme(style="whitegrid")

# Create figure
fig, axs = plt.subplots(1, 2, figsize=(14, 5))

# 1. Distribution Plot (Kernel Density Estimate & Histogram)
sns.histplot(data=tips, x="total_bill", kde=True, ax=axs[0], color="purple")
axs[0].set_title('Total Bill Distribution')

# 2. Boxplot (Categorical comparison)
sns.boxplot(data=tips, x="day", y="total_bill", hue="sex", ax=axs[1], palette="Set2")
axs[1].set_title('Total Bill by Day & Sex')

plt.tight_layout()
plt.show()

```

## 6.2 Matrix Plots (Heatmaps) and Pair Plots
Matrix plots are crucial for visualizing correlation matrices during feature analysis.



```python
# Chapter 6.2: Heatmaps & Pairplots
# Calculate Correlation Matrix
corr = tips[['total_bill', 'tip', 'size']].corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".3f", linewidths=.5)
plt.title("Feature Correlation Matrix")
plt.show()

# Pairplot (Pairwise relationships across features)
sns.pairplot(tips[['total_bill', 'tip', 'sex']], hue='sex', palette='husl')
plt.show()

```

## 7.1 Linear Algebra: Vectors, Matrices, and Eigendecomposition
Eigendecomposition ($Av = \lambda v$) is core to dimensionality reduction algorithms like PCA.



```python
# Chapter 7.1: Matrix Operations & Eigenvalues
import numpy as np

# Define a symmetric matrix (e.g. Covariance matrix style)
A = np.array([[4, 2],
              [2, 3]])

# Calculate Eigenvalues and Eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

print("Matrix A:
", A)
print("
Eigenvalues (λ):", eigenvalues)
print("
Eigenvectors (v) [each column represents an eigenvector]:
", eigenvectors)

# Verification: A * v = λ * v
v1 = eigenvectors[:, 0]
lambda1 = eigenvalues[0]
lhs = A @ v1
rhs = lambda1 * v1
print(f"
Verification for first Eigenpair:")
print(f"A * v1 = {lhs}")
print(f"λ1 * v1 = {rhs}")
print(f"Are they close? {np.allclose(lhs, rhs)}")

```

## 7.2 Probability & Statistics: Bayes Theorem
Bayes Theorem calculates conditional probability:
$$P(A|B) = rac{P(B|A) \cdot P(A)}{P(B)}$$

### Real-World Example
Determining the probability that a patient has a disease given a positive lab test result.
- $P(D)$: Prior probability of having the disease ($1\%$).
- $P(T|D)$: Sensitivity of test ($99\%$).
- $P(T|D^c)$: False positive rate ($5\%$).



```python
# Chapter 7.2: Bayes Theorem Simulation
# Probabilities
p_disease = 0.01  # P(D)
p_no_disease = 0.99  # P(D^c)

p_pos_given_disease = 0.99  # P(T|D) - Sensitivity
p_pos_given_no_disease = 0.05  # P(T|D^c) - False Positive Rate

# Total probability of testing positive: P(T) = P(T|D)*P(D) + P(T|D^c)*P(D^c)
p_pos = (p_pos_given_disease * p_disease) + (p_pos_given_no_disease * p_no_disease)

# Posterior probability: P(D|T) = P(T|D)*P(D) / P(T)
p_disease_given_pos = (p_pos_given_disease * p_disease) / p_pos

print(f"Prior Probability of Disease: {p_disease * 100}%")
print(f"Total Probability of testing Positive: {p_pos * 100:.2f}%")
print(f"Posterior Probability (Actual Disease given Positive Test): {p_disease_given_pos * 100:.2f}%")

```

## 7.3 Optimization: Gradient Descent
Gradient descent is the primary optimization algorithm used to minimize loss functions in machine learning.
$$w \leftarrow w - \eta rac{\partial L}{\partial w}$$
where $\eta$ is the learning rate.

### Code: 1D Gradient Descent Visualization



```python
# Chapter 7.3: Gradient Descent Implementation
# Define Objective function: f(x) = x^2 - 4x + 6
# Derivative: f'(x) = 2x - 4

def f(x):
    return x**2 - 4*x + 6

def df(x):
    return 2*x - 4

# Gradient Descent setup
learning_rate = 0.1
x_start = 0.0
iterations = 20

x_path = [x_start]
y_path = [f(x_start)]

x = x_start
for i in range(iterations):
    grad = df(x)
    x = x - learning_rate * grad
    x_path.append(x)
    y_path.append(f(x))

# Plotting the path
x_grid = np.linspace(-1, 5, 100)
plt.figure(figsize=(8, 5))
plt.plot(x_grid, f(x_grid), 'k-', label='Loss Surface: $f(x) = x^2 - 4x + 6$')
plt.scatter(x_path, y_path, color='red', marker='o', label='GD Steps')
plt.plot(x_path, y_path, color='red', linestyle=':', alpha=0.6)
plt.title('Gradient Descent Optimization Path')
plt.xlabel('Parameter w (or x)')
plt.ylabel('Loss Value')
plt.legend()
plt.grid(True)
plt.show()

print(f"Optimized parameter value: {x:.6f} (Analytical minimum is 2.000000)")

```
