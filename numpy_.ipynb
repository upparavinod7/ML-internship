```python
x = np.linspace(-10,10,100)
def Relu(x):
    return np.maximum(0,x)
Y = Relu(x)
plt.plot(x,Y)
plt.grid(True)
plt.show()

```

```python
x = np.linspace(-10,10,100)
def sigmoid(x):
    return 1/1+np.exp(-x)
Y = sigmoid(x)
plt.grid(True)
plt.show()
```

```python
year = range(2000,2008)
apple_yield = [0.85, 0.89, 0.91, 0.93, 0.92, 0.95, 0.97, 0.96]
orange_yield = [0.72, 0.75, 0.78, 0.81, 0.80, 0.83, 0.84, 0.87]
plt.figure(figsize = (8,6))
plt.plot(year, apple_yield)
plt.plot(year, orange_yield)
plt.legend(['apples', 'oranges'])
plt.xlabel('Year')
plt.ylabel('Yield (Tons)')
plt.title("Yield in Fruits")
plt.show()
```
