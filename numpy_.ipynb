```python
x = np.linspace(-10,10,100)
def Relu(x):
    return np.maximum(0,x)
Y = Relu(x)
plt.plot(x,Y)
plt.grid(True)
plt.show()

```

```python
x = np.linspace(-10,10,100)
def sigmoid(x):
    return 1/1+np.exp(-x)
Y = sigmoid(x)
plt.plot(x,Y)
plt.grid(True)
plt.show()
```

```python
def Relu(x):
    return np.maximum(0,x)
x = np.linspace(-10,10, 100)
y = Relu(x)
plt.plot(x,y)
```

```python
def binaryStep(x):
    return np.where(x >= 0, 1, 0)
x = np.linspace(-10,10, 100)
y = binaryStep(x)
plt.plot(x,y)
```

```python
def LinearActivation(x):
    return x
x = np.linspace(-10,10, 100)
y = LinearActivation(x)
plt.plot(x,y)
```

```python
def LeakyRelu(x):
    return np.where(x>0, 1, 0.01*x)
x = np.linspace(-10,10, 100)
y = LeakyRelu(x)
plt.plot(x,y)
```

```python
def swish(x):
    return sigmoid(x)*x
x = np.linspace(-10,10, 100)
y = swish(x)
plt.plot(x,y)
```
